```python
import pandas as pd
```

```python
dataset = pd.read_csv('taipei_city_real_estate_transaction_v2.csv')
dataset.head()
```

```python
# One hot encoding for transaction type
# Add land, building, garage column
dataset = dataset.assign(land = 0, building = 0, garage = 0)
# Analysis transaction type
for index in range(len(dataset)):
    if(dataset.loc[index, 'transaction_type'] == 'Land'):
        dataset.loc[index, 'land'] = 1
    elif(dataset.loc[index, 'transaction_type'] == 'Building'):
        dataset.loc[index, 'building'] = 1
    elif(dataset.loc[index, 'transaction_type'] == 'Garage'):
        dataset.loc[index, 'garage'] = 1
    elif(dataset.loc[index, 'transaction_type'] == 'Premises (land + building)'):
        dataset.loc[index, 'land'] = 1
        dataset.loc[index, 'building'] = 1
    elif(dataset.loc[index, 'transaction_type'] == 'Premises (land + building) + parking space'):
        dataset.loc[index, 'land'] = 1
        dataset.loc[index, 'building'] = 1
        dataset.loc[index, 'garage'] = 1
    else:
        print("!!!ERROR!!!")
dataset = dataset.drop(['transaction_type'], axis = 1)
dataset.head()
```

```python
# One hot encoding for urban land use
urbanLandUse = pd.get_dummies(dataset.urban_land_use)
dataset = dataset.merge(urbanLandUse, left_index = True, right_index = True)
dataset = dataset.drop(['urban_land_use'], axis = 1)
# One hot encoding for main use
mainUse = pd.get_dummies(dataset.main_use)
dataset = dataset.merge(mainUse, left_index = True, right_index = True)
dataset = dataset.drop(['main_use'], axis = 1)
# One hot encoding for main building material
mainBuildingMaterial = pd.get_dummies(dataset.main_building_material)
dataset = dataset.merge(mainBuildingMaterial, left_index = True, right_index = True)
dataset = dataset.drop(['main_building_material'], axis = 1)
# One hot encoding for carpark category
carparkCategory = pd.get_dummies(dataset.carpark_category)
dataset = dataset.merge(carparkCategory, left_index = True, right_index = True)
dataset = dataset.drop(['carpark_category'], axis = 1)
dataset.head()
```

```python
# Change number of partition & management organization type
for index in range(len(dataset)):
    if(dataset.loc[index, 'num_partition'] == 'æœ‰'):
        dataset.loc[index, 'num_partition'] = 1
    else:
        dataset.loc[index, 'num_partition'] = 0
    if(dataset.loc[index, 'management_org'] == 'æœ‰'):
        dataset.loc[index, 'management_org'] = 1
    else:
        dataset.loc[index, 'management_org'] = 0
dataset.head()
```

```python
dataset.to_csv('./District/BeforeSeparateDistrict.csv')
```
