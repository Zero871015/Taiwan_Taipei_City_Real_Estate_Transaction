```python
import pandas as pd
```

```python
dataset = pd.read_csv('taipei_city_real_estate_transaction_v2.csv')
dataset.head()
```

```python
# Delete no relation column
dataset = dataset.drop(['transaction_type', 'urban_land_use', 'main_use', 'main_building_material', 'carpark_category'], axis = 1)
dataset.head()
```

```python
# Change number of partition & management organization type
for index in range(len(dataset)):
    if(dataset.loc[index, 'num_partition'] == '有'):
        dataset.loc[index, 'num_partition'] = 1
    else:
        dataset.loc[index, 'num_partition'] = 0
    if(dataset.loc[index, 'management_org'] == '有'):
        dataset.loc[index, 'management_org'] = 1
    else:
        dataset.loc[index, 'management_org'] = 0
dataset.head()
```

```python
# Numbering district
for index in range(len(dataset)):
    if(dataset.loc[index, 'district'] == 'Daan District'):
        dataset.loc[index, 'district'] = 12
    elif(dataset.loc[index, 'district'] == 'Xinyi District'):
        dataset.loc[index, 'district'] = 11
    elif(dataset.loc[index, 'district'] == 'Zhongzheng District'):
        dataset.loc[index, 'district'] = 10
    elif(dataset.loc[index, 'district'] == 'Songshan District'):
        dataset.loc[index, 'district'] = 9
    elif(dataset.loc[index, 'district'] == 'Zhongshan Area'):
        dataset.loc[index, 'district'] = 8
    elif(dataset.loc[index, 'district'] == 'shihlin'):
        dataset.loc[index, 'district'] = 7
    elif(dataset.loc[index, 'district'] == 'Neihu District'):
        dataset.loc[index, 'district'] = 6
    elif(dataset.loc[index, 'district'] == 'Datong District'):
        dataset.loc[index, 'district'] = 5
    elif(dataset.loc[index, 'district'] == 'Nangang District'):
        dataset.loc[index, 'district'] = 4
    elif(dataset.loc[index, 'district'] == 'Beitou'):
        dataset.loc[index, 'district'] = 3
    elif(dataset.loc[index, 'district'] == 'Wanhua District'):
        dataset.loc[index, 'district'] = 2
    elif(dataset.loc[index, 'district'] == 'Wenshan District'):
        dataset.loc[index, 'district'] = 1
dataset.head()
```

```python
dataset.to_csv('./District/NumberingDistrict.csv', index = False)
```
